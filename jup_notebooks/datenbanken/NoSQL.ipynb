# NoSQL

## Abgrenzung zu anderen Datenbanksystem


Next Generation Databases mostly addressing some of the points: 
being non-relational,
distributed, open-source and horizontal scalable. 
The original intention has been modern web-scale databases. [...] 

Often more characteristics apply as: schema-free, easy replication support, simple API, eventually consistent / BASE (not ACID), and more. So the misleading term
” nosql“ [...] should be seen as an alias to something like the deﬁnition above.
