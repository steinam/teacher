Schleifen
=====

Stehen fr wiederholte Anweisungen



```python
n = 5
while n > 0:
    print(n)
    n = n - 1
print("Blastoff!")
print(n)
```

Eine Schleife verlassen
-------------------------



```python
while True:
    line = input('> ')
    if line == 'done' :
        break
    print(line)
print("Done!")
```
